# 2. Check the question type
## a. Shape, Material, and Purpose
Require knowledge of basic properties of the objects (flexibility, curvature, and being porous)

1. Make an outdoor pillow
  - Blow into a **tin can** and tie with rubber band
  - <u>Blow into a **trash bag** and tie with rubber band</u>

2. To make a hard shelled taco,
  - put seasoned beef, cheese, and lettuce **onto** the hard shell.
  - <u>put seasoned beef, cheese, and lettuce **into** the hard shell.</u>

3. How do I find something I lost on the carpet?
  - Put a **solid seal** on the end of your vacuum and turn it on.
  - <u>Put a **hair net** on the end of your vacuum and turn it on.</u>

## b. Commonsense Convenience
Both answers may be technically correct, but one is more convenient and preferable
1. How to make sure all the clocks in the house are set accurately?
  - Get a solar clock for a reference and place it just outside a window that gets lots of sun. Use a system of call and response once a month, having one person stationed at the solar clock who yells out the correct time and have another person move to each of the indoor clocks to check if they are showing the right time. Adjust as necessary.
  - <u>Replace all wind-ups with digital clocks. That way, you set them once, and that’s it. Check the batteries once a year or if you notice anything looks a little off.</u>

````python
from difflib import SequenceMatcher

for row in train.itertuples():
    if SequenceMatcher(None, row.sol1, row.sol2).ratio() > 0.5:
        train.at[row.Index, 'type'] = 'a'
    else:
        train.at[row.Index, 'type'] = 'b'

display(train.head())
````

````python
from tqdm.notebook import tqdm

nlp = spacy.load('en_core_web_sm')

def findCorrects(correct, wrong):
    correctList = correct.copy()

    for i in range(len(wrong)):
        for j in range(len(correctList)):
            if wrong[i] == correctList[j]:
                correctList.pop(j)
                break
    
    if correctList == []:
        return 'null'

    return ','.join(correctList)

def findWrongs(correct, wrong):
    wrongList = wrong.copy()

    for i in range(len(correct)):
        for j in range(len(wrongList)):
            if correct[i] == wrongList[j]:
                wrongList.pop(j)
                break

    if wrongList == []:
        return 'null'
    
    return ','.join(wrongList)

for row in tqdm(train.itertuples()):
    if row.type == 'a':
        ans = row.answer
        col1 = nlp(row.sol1)
        col2 = nlp(row.sol2)

        nounChunks1 = [chunk.lemma_ for chunk in col1.noun_chunks]
        nounChunks2 = [chunk.lemma_ for chunk in col2.noun_chunks]
        verbs1  = [token.lemma_ for token in col1 if token.pos_ == 'VERB']
        verbs2  = [token.lemma_ for token in col2 if token.pos_ == 'VERB']
    
        if answer == '0':
            train.at[row.Index, 'correctNounChunks'] = findCorrects(nounChunks1, nounChunks2)
            train.at[row.Index, 'wrongNounChunks'] = findWrongs(nounChunks1, nounChunks2)
            train.at[row.Index, 'correctVerbChunks'] = findCorrects(verbs1, verbs2)
            train.at[row.Index, 'wrongVerbChunks'] = findWrongs(verbs1, verbs2)
        
        else: # answer == '1':
            train.at[row.Index, 'correctNounChunks'] = findCorrects(nounChunks2, nounChunks1)
            train.at[row.Index, 'wrongNounChunks'] = findWrongs(nounChunks2, nounChunks1)
            train.at[row.Index, 'correctVerbChunks'] = findCorrects(verbs2, verbs1)
            train.at[row.Index, 'wrongVerbChunks'] = findWrongs(verbs2, verbs1)
            
````
